1. If the value for `user_input_month` is less than 1 or greater than 12, set `valid_month` to `False` and print out `"Invalid month. Please enter a number between 1-12."` Otherwise, set `valid_month` to `True`.

However, later we'll be looping through these `if...else` statements multiple times, which is why this step is necessary.*

````python
# Copy and paste your code from the previous cell and add then modify it below:
valid_month = False
valid_day = False


def user_input_validation(user_input_month,user_input_day):
    if user_input_month < 1 or user_input_month > 12:
        valid_month = False
        print ("Invalid month. Please enter a number between 1 and 12")
        
    if user_input_day < 1 or user_input_day > 31:
        print ("Invalid day. Please enter a number between 1 and 31")
        valid_day = False
    if user_input_day == True and user_input_month == True:
        return True
    return True
````
